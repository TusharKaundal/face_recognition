```python
import cv2
import os
import numpy as np
import shutil 
path='C://Users//TUSHAR//Desktop//projects//faces_dataset//'
shutil.rmtree(path, ignore_errors=True)
new_path_test ='C://Users//TUSHAR//Desktop//projects//faces_dataset//test//'
os.makedirs(new_path_test)
new_path_train='C://Users//TUSHAR//Desktop//projects//faces_dataset//train//'
os.makedirs(new_path_train)
# Load HAAR face classifier
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Load functions
def face_extractor(img):
    # Function detects faces and returns the cropped face
    # If no face detected, it returns the input image
    
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    
    if faces is ():
        return None
    
    # Crop all faces found
    for (x,y,w,h) in faces:
        cropped_face = img[y:y+h, x:x+w]

    return cropped_face

# Initialize Webcam
person = ['Krish','Kapil']
for x in person:
    cap = cv2.VideoCapture(0)
    count = 0
    new_path_train1 = os.path.join(new_path_train, x) 
    os.mkdir(new_path_train1)
    new_path_test1 = os.path.join(new_path_test, x) 
    os.mkdir(new_path_test1)

    # Collect 150 samples of your face from webcam input
    while True:

        ret, frame = cap.read()
        if face_extractor(frame) is not None:
            count += 1
            face = cv2.resize(face_extractor(frame), (200, 200))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
            # Save file in train directory with unique name
            if count <= 100 :
                file_name_path = new_path_train1 + '//face' + str(count) + '.jpg'
                cv2.imwrite(file_name_path, face)

                # Put count on images and display live count
                cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
                cv2.imshow('Face Cropper', face)
            else:
                counts=count-100
                # Save file in train directory with unique name
                file_name_path = new_path_test1 + '//face' + str(counts) + '.jpg'
                cv2.imwrite(file_name_path, face)

                # Put count on images and display live count
                cv2.putText(face, str(counts), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
                cv2.imshow('Face Cropper', face)

        else:
            print("Face not found")
            pass

        if cv2.waitKey(1) == 13 or count == 150: #13 is the Enter Key
            break

    cap.release()
    cv2.destroyAllWindows()      
print("Collecting Samples Complete")
```
